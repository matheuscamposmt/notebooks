# Linear Regression

Linear regression is a linear approach for modelling the relationship between a scalar response and one or more explanatory variables (also known as dependent and independent variables).

## Physics and History grades example:
Suppose you study at the Data Science High School and your grades in Physics and History are the following:

**Physics Scores**:  15  12  8   8   7   7   7   6   5   3

**History Scores**:  10  25  17  11  13  17  20  13  9   15

Then, you wonder if there is a relationship between your scores in Physics and History. To build a relationship model you can use a **linear** function and try to fit the best line when Physics is the $x$ variable and History is the $y$ variable.

```python
x = [15,12,8,8,7,7,7,6,5,3]
y = [10,25,17,11,13,17,20,13,9,15]
xy = [x_i * y_i for x_i, y_i in zip(x,y)]
x2 = [x_i**2 for x_i in x]
n = len(x)

def slope(x, y):
    num = n * sum(xy) - sum(x) * sum(y)
    den = n * sum(x2) - (sum(x))**2
    slope = num/den

    return round(slope, 3)

def intercept(x, y):
    a = slope(x, y)
    b = (sum(y) - a * sum(x)) / n
    
    return b

a = slope(x,y)
b = intercept(x,y)

print(f"y = {a:.3f} * x + {b:.3f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_linear_function(a, b, x=None, y=None):
    xs = np.linspace(0, 16)
    ys = a * xs + b

    plt.plot(xs, ys, color='black')
    if x and y:
        plt.scatter(x, y, color='blue', label='Experimental points')
    
    plt.title('Linear Regression')
    plt.legend()
```

```python
plot_linear_function(a, b, x=x, y=y)
```
